```python
#Array Creation
#These functions are used to initialize new array objects:
np.array(): #Creates an array from a Python list or tuple.
np.zeros() / np.ones(): #Creates arrays filled with 0s or 1s of a specified shape.
np.arange(start, stop, step): #Returns evenly spaced values within a given interval (similar to Python's range).
np.linspace(start, stop, num): #Returns a specific number of evenly spaced numbers over a range.
np.eye(): #Creates an identity matrix (useful in linear algebra).
```

```python
#Array Manipulation
#Used to change the structure of existing arrays without altering their data:
np.reshape(): #Changes the shape of an array (e.g., converting a 1D array to a 2D matrix).
np.transpose() or #.T:# Reverses the axes of an array (swaps rows and columns).
np.concatenate(): #Joins two or more arrays along a specified axis.
np.vstack() / np.hstack(): #Stacks arrays vertically or horizontally.
np.flatten() / np.ravel(): #Collapses a multi-dimensional array into a single dimension (1D)
```

```python
#Mathematical & Statistical Functions
#NumPy performs these operations element-wise, which is significantly faster than using loops:
Aggregations: #np.sum()#, np.mean(), np.median(), np.std() (standard deviation), and np.var() (variance).
Extrema: #np.min(), np.max(), np.argmin(), and np.argmax() (returns the index of the min/max value).
Arithmetic:# np.add(), np.subtract(), np.multiply(), np.divide(), and np.sqrt().
Trigonometry: #np.sin(), np.cos(), np.tan(), and their inverses.
```

```python
#Logic, Searching, & Sorting
np.where(): #Returns indices where a specific condition is met; essentially a vectorized "if-else".
np.sort(): #Returns a sorted copy of an array
np.unique(): #Finds all unique elements in an array
```
